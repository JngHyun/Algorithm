```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = CountVectorizer(min_df=1,ngram_range=(1,1))
```

```python
corpus = final
```

```python
features = vectorizer.fit_transform(corpus)
```

```python
dtm_np = np.array(features.todense())
print(dtm_np)
```

```python
feature_names = vectorizer.get_feature_names()
print(feature_names)
```

```python
pd.DataFrame(data=dtm_np,columns = feature_names)
```

```python
eds = []
for k in range(len(corpus)):
    for i in range(len(corpus)):
        ed = np.linalg.norm(dtm_np[k]-dtm_np[i])
        eds.append(ed)
        print(ed)
```

```python
ud_np = np.array(eds)
ud_np.reshape(len(corpus),len(corpus))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig = plt.figure(figsize=(8,8))
sns.heatmap(data=ud_np.reshape(len(corpus),len(corpus)),annot=True,cmap='summer')
```

```python
# 모든 코사인 유사도 구하기
cosine_similarity = []
for k in range(len(corpus)):
    for i in range(len(corpus)):
        cosine = np.dot(dtm_np[k],dtm_np[i])/(np.linalg.norm(dtm_np[k])*np.linalg.norm(dtm_np[i]))
        cosine_similarity.append(cosine)
        print(cosine)
```

```python
cs_np = np.array(cosine_similarity)
cs_np.reshape(len(corpus),len(corpus))
```

```python
fig = plt.figure(figsize=(8,8))
sns.heatmap(data = cs_np.reshape(len(corpus),len(corpus)), annot = True, fmt = '.2f', linewidths=.5)
```

```python
df = pd.read_csv('./data/영화 시놉시스 데이터.csv', index_col = 0)
df = df.iloc[:100,:]
```

```python
import re
def extra_process(x):
    x = re.sub("[^\w\s]"," ",x) # 모든 특수문자 제거
    x = re.sub("접기"," ",x) # "접기"라는 단어
    return x
```

```python
df['시놉시스'] = df['시놉시스'].apply(extra_process)
```

```python
df.isnull().sum()
```

```python
df[df.duplicated(subset='영화제목')]
```

```python
df.drop_duplicates(subset='영화제목',keep='first',inplace=True)
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer()
features = vectorizer.fit_transform(df['시놉시스'])
```

```python
feature_names = vectorizer.get_feature_names()
```

```python
dtm_np = np.array(features.todense())
```

```python
pd.DataFrame(data= dtm_np, columns =feature_names)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cos_sim = cosine_similarity(dtm_np,dtm_np)
```

```python
indices = pd.Series(df.index,index=df['영화제목'])
```

```python
indices
```

```python
idx=indices['모아나']
```

```python
sim_scores = list(enumerate(cos_sim[idx]))
```

```python
sim_scores = sorted(sim_scores,key= lambda x: x[1],reverse=True)
```

```python
book_indices = [i[0] for i in sim_scores][:10]
```

```python
print(df['영화제목'].iloc[book_indices])
```
